```python
def pandas_merger(ruta):
    path = ruta
    proyecto_sn_xlsx = ruta[0:-5]
    
    
    #Data frame del archivo de Excel
    
    df_reporte = pd.read_excel(path,header = 2)
    x = len(df_reporte)
    
    #Borrar las primeras dos columnas
    
    df_reporte = df_reporte.drop([x-1])
    df_reporte = df_reporte.drop(["Unnamed: 0","Unnamed: 1"], axis = 1)
    
    #Generas una lista con el nombre de las columnas del data frame y generas de ella una lista de las columnas numéricas para su limpieza. 

    col_a_lista = list(df_reporte)
    col_a_lista.remove('N° Crédito')
    col_num = col_a_lista[8:]
    
    #Se pasan las columnas numéricas a tipo de datos 'string', para poderlas limpiar con la función 'replace' quitando las comas, los signos de pesos y los porcentajes, para despúes tranformar estos datos a tipo 'float64'

    #df_reporte['N° Crédito'] = df_reporte['N° Crédito'].astype('int')
    df_reporte = df_reporte.astype('string')
    
    for col in col_a_lista:
        df_reporte[col] = df_reporte[col].str.replace('$','')
        df_reporte[col] = df_reporte[col].str.replace('%','')
        df_reporte[col] = df_reporte[col].str.replace(',','')
        
    df_reporte[col_num] = df_reporte[col_num].astype('float64')
    
    #Se obtienen las listas de las columnas a sumar y eliminar
    
    mins_netas = []
    for col in col_a_lista:
        for i in range(len(col)):
            if col[i:i+14] == "Ministración N" or col[1:2] == "An" :
                mins_netas.append(col)

    cobro_comision = []
    for col in col_a_lista:
        for i in range(len(col)):
            if col[i:i+20] == "Comisión Mercantil y":
                cobro_comision.append(col)

    avance_parcial = []
    for col in col_a_lista:
        for i in range(len(col)):
            if col[i:i+14] == "Avance parcial":
                avance_parcial.append(col)

    fg_ret = []
    for col in col_a_lista:
        for i in range(len(col)):
            if col[i:i+6] == "Fondo ":
                fg_ret.append(col)
                
    #Se realizan las sumas de las columnas de ministraciones, cobro de comisiones, retención de fg, avance parcial, remanente en cta. y min. neta/tot. de crédito
    
    serie_mins_netas = round(df_reporte[mins_netas].sum(axis = 1), 2)
    serie_cobro_comision = abs(round(df_reporte[cobro_comision].sum(axis = 1), 2))
    serie_avance_parcial = round(df_reporte[avance_parcial].sum(axis = 1), 2)
    serie_fg_ret = round(df_reporte[fg_ret].sum(axis = 1), 2)
    
    #Se revisa si ya se pago el fondo de garantía.
    
    if proyecto_sn_xlsx in lfg:
        serie_min_neta_tot_cred = round(((serie_mins_netas + serie_fg_ret + df_reporte['Anticipo Neto para el constructor'] + serie_cobro_comision) / df_reporte['Total crédito']) * 100,2)
        serie_remanente_cuenta = round(df_reporte['Total crédito'] - (serie_mins_netas + serie_cobro_comision + df_reporte['Anticipo Neto para el constructor'] + serie_fg_ret), 2)
    else:
        serie_min_neta_tot_cred = round(((serie_mins_netas + df_reporte['Anticipo Neto para el constructor'] + serie_cobro_comision) / df_reporte['Total crédito']) * 100,2)
        serie_remanente_cuenta = round(df_reporte['Total crédito'] - (serie_mins_netas + serie_cobro_comision + df_reporte['Anticipo Neto para el constructor']), 2)
    
    #Se genera un nuevo df para con las columnas de las operaciones
    
    df_reporte2 = pd.DataFrame({"Avance de obra (%)":serie_avance_parcial,
                            "Ministración pagada acum.":serie_mins_netas,
                           "Comisión cobrada acum.":serie_cobro_comision,
                           "Fondo de garantía ret.": serie_fg_ret,
                           "Remanente en cta.":serie_remanente_cuenta,
                           "Min neta/Tot. cred. (%)":serie_min_neta_tot_cred})
    
    #Se genera un nuevo df (df_final) concatenando df_reporte y df_reporte2
    
    df_final = pd.concat([df_reporte,df_reporte2],axis = 1)
    
    #Se remueven de la lista de col_num 'Total crédito','Anticipo Neto para el constructor','Saldo del proyecto'
    
    col_num.remove('Total crédito')
    col_num.remove('Anticipo Neto para el constructor')
    col_num.remove('Saldo del proyecto')
    
    #Se realiza drop a las columnas en la lista col_num
    
    df_final = df_final.drop(columns = col_num)
    
    #Se realiza un merge al df_base_dae
    
    return df_final
```

```python
#Se generan solo las columnas del Data Frame df_base_dae 
    
df_base_dae = pd.DataFrame(columns = ['Proyecto','Etapa','Paquete','Acreditado',
 'N° Crédito',
 'NSS',
 'Calle',
 'Mza.',
 'Lote',
 'Num Ext.',
 'Num Int.',
 'M2 Terreno Vendible',
 'Total crédito',
 'Anticipo Neto para el constructor',
 'Saldo del proyecto',
 'Avance de obra (%)',
 'Ministración pagada acum.',
 'Comisión cobrada acum.',
 'Fondo de garantía ret.',
 'Remanente en cta.',
 'Min neta/Tot. cred. (%)'])
```

```python
#Se obtiene la lista de las rutas

lista_reportes = os.listdir()
```

```python
#Los archivos descargados de sistema tienen un error en el RGB set por lo que al tratar de manipularlos con pandas te arroja un error, con estas lpineas de código se arregla esto 

__old_rgb_set__ = RGB.__set__


def __rgb_set_fixed__(self, instance, value):
    try:
        __old_rgb_set__(self, instance, value)
    except ValueError as e:
        if e.args[0] == 'Colors must be aRGB hex values':
            __old_rgb_set__(self, instance, WHITE)  # Change default color here


RGB.__set__ = __rgb_set_fixed__
```

```python
#Se hace un loop con la función de pandas merge para transofrmar y unir los data frames generados de cada archivo en un data frame final llamado df_base_dae, en este loop se agregan las columnas de Proyecto, Paquete y etapa
# a cada archivo individual

for reporte in lista_reportes:
    a = pandas_merger(reporte)
    lista = reporte.split("-")
    lista = lista[0:3]
    lista1 =[]
    ii = 0
    while ii < len(a):
        lista1.append(lista)
        ii+=1
    b = pd.DataFrame(np.array(lista1),columns = ['Proyecto','Etapa','Paquete'])
    c = pd.concat([b,a],axis = 1)
    df_base_dae = df_base_dae.merge(c,how = 'outer')
    
#df_base_dae = df_base_dae.dropna()
```
